```python
import pandas as pd
import matplotlib.pyplot as plt
```

Let's organize the data first. \
First thing first set Date as the index.

```python
df = pd.read_csv('data/CLMMAXT_HKO_.csv')
df = df[df['data Completeness'] == 'C'] # only keep 'good' data

df['Month'] = df['Month'].astype(str).str.zfill(2)
df['Day'] = df['Day'].astype(str).str.zfill(2)
df['Year'] = df['Year'].astype(str)
df['Date'] = df['Year'] + '-' + df['Month'] + '-' + df['Day']

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Temperature'] = pd.to_numeric(df['Value'], errors='coerce') #it somehow tread those number in the csv as str =.=

df.set_index('Date', inplace=True)
df.drop(['Year', 'Month', 'Day','Value','data Completeness'], axis=1, inplace=True)
```

```python
df['Temperature'].plot()
```
