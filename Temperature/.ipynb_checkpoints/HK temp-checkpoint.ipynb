```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
import torch
import torch.nn as nn
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

Let's organize the data first. \
First thing first set Date as the index.

```python
df = pd.read_csv('data/CLMMAXT_HKO_.csv')
df = df[df['data Completeness'] == 'C'] # only keep 'good' data

df['Month'] = df['Month'].astype(str).str.zfill(2)
df['Day'] = df['Day'].astype(str).str.zfill(2)
df['Year'] = df['Year'].astype(str)
df['Date'] = df['Year'] + '-' + df['Month'] + '-' + df['Day']

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Temperature'] = pd.to_numeric(df['Value'], errors='coerce') #it somehow tread those number in the csv as str =.=

df.set_index('Date', inplace=True)
df.drop(['Year', 'Month', 'Day','Value','data Completeness'], axis=1, inplace=True)
```

Temperature prediction

```python
prophet_df = df['Temperature'].reset_index()
prophet_df.columns = ['ds', 'y']

prophet_model = Prophet(yearly_seasonality=True)
prophet_model.fit(prophet_df)
```

```python
future = prophet_model.make_future_dataframe(periods=365)  # predict next year
forecast = prophet_model.predict(future)
```

```python
class LSTMPredictor(nn.Module):
    def __init__(self, input_size=1, hidden_size=50, num_layers=1):
        super(LSTMPredictor, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.linear = nn.Linear(hidden_size, 1)
    
    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        predictions = self.linear(lstm_out[:, -1, :])
        return predictions

def prepare_data(df, seq_length=60, train_split=0.8):
    
    def create_sequences(data, seq_length):
        X, y = [], []
        for i in range(len(data) - seq_length):
            X.append(data[i:i+seq_length])
            y.append(data[i+seq_length])
        return np.array(X), np.array(y)
    
    # Scale the data
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(df['Temperature'].values.reshape(-1, 1))
    
    # Create sequences
    X, y = create_sequences(scaled_data, seq_length)
    
    # Convert to PyTorch tensors
    X = torch.FloatTensor(X)
    y = torch.FloatTensor(y)
    
    # Split into train and test sets
    train_size = int(len(X) * train_split)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]
    
    print(f"Training data shape: {X_train.shape}")
    print(f"Test data shape: {X_test.shape}")
    
    return {
        'X_train': X_train,
        'y_train': y_train,
        'X_test': X_test,
        'y_test': y_test,
        'scaler': scaler
    }

def build_and_train(prepared_data, batch_size=32, num_epochs=50, learning_rate=0.01):
    
    X_train = prepared_data['X_train']
    y_train = prepared_data['y_train']
    
    # Initialize model and training components
    model = LSTMPredictor()
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
    
    # Training Loop with progress monitoring
    losses = []
    
    print("\nStarting training...")
    for epoch in range(num_epochs):
        model.train()
        epoch_losses = []
        
        for i in range(0, len(X_train), batch_size):
            batch_X = X_train[i:i+batch_size]
            batch_y = y_train[i:i+batch_size]
            
            optimizer.zero_grad()
            outputs = model(batch_X)
            loss = criterion(outputs, batch_y)
            loss.backward()
            optimizer.step()
            
            epoch_losses.append(loss.item())
        
        avg_loss = sum(epoch_losses) / len(epoch_losses)
        losses.append(avg_loss)

        if (epoch+1)%10 == 0:
            print(f'Epoch [{epoch+1}/{num_epochs}] - Loss: {avg_loss:.4f}')
        
        # Early stopping check (optional)
        if avg_loss < 1e-4:
            print("Loss threshold reached, stopping training")
            break
    
    
    # Plot training loss
    plt.figure(figsize=(10, 5))
    plt.plot(losses)
    plt.title('Training Loss Over epoch')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.show()
    
    return {
        'model': model,
        'training_loss': losses,
    }

def predict(model, scaler, input_data, future_steps=30, plot=True):
    
    model.eval()
    future_predictions = []
    current_sequence = input_data.clone() if isinstance(input_data, torch.Tensor) else torch.FloatTensor(input_data)
    
    print("Making predictions...")
    with torch.no_grad():
        for step in range(future_steps):
            prediction = model(current_sequence.unsqueeze(0))
            future_predictions.append(prediction.item())
            
            # Update sequence for next prediction
            current_sequence = torch.cat((current_sequence[1:], prediction), dim=0)
            
            if (step + 1) % 10 == 0:
                print(f"Completed {step + 1}/{future_steps} predictions")
    
    # Inverse transform predictions
    future_predictions = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))
    
    if plot:
        plt.figure(figsize=(12, 6))
        plt.plot(future_predictions, label='Predictions')
        plt.title('Temperature Predictions')
        plt.xlabel('Time Steps')
        plt.ylabel('Temperature')
        plt.legend()
        plt.show()
    
    return future_predictions
```

```python
prepared_data = prepare_data(df)
training_results = build_and_train(prepared_data)
model = training_results['model']
```

```python
last_sequence = prepared_data['X_test'][-1]
predictions = predict(model, prepared_data['scaler'], last_sequence,365)
```
